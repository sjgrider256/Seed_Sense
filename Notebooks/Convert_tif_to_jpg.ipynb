```python
import os
import glob
import rasterio
import numpy as np
from PIL import Image
from pathlib import Path
```

```python
#Set paths
tile_dir = os.path.join("Data", "Tiles")
```

```python
# Convert TIF files to JPG
converted = 0

for tif_path in glob.glob(os.path.join(tile_dir, "*.tif")):
    jpg_path = os.path.splitext(tif_path)[0] + ".jpg"

    try:
        with rasterio.open(tif_path) as src:
            img = src.read([1, 2, 3])  # first three bands (RGB). removes 4th band from DJI
            img = np.transpose(img, (1, 2, 0))  # reshape to (height, width, channels)
            img = np.clip(img, 0, 255).astype("uint8")

        Image.fromarray(img).save(jpg_path, "JPEG", quality=90)
        converted +=1
    except Exception as e:
        print(f"Failed on {os.path.basename(tif_path)}: {e}")

print(f"Done. Converted {converted} TIFs to JPG.")
```
